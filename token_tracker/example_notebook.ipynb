# Token Tracker - Exemple d'utilisation dans Jupyter Lab

Ce notebook montre comment utiliser le Token Tracker pour suivre l'utilisation des tokens par requête.

```python
# Installation (si nécessaire)
# !pip install pandas matplotlib
```

```python
# Importation
from token_tracker import TokenTracker
import pandas as pd
```

## 1. Initialisation du Tracker

```python
# Créer un tracker avec un fichier de stockage
tracker = TokenTracker("mon_suivi_tokens.json")

# Afficher l'état initial
print(tracker)
```

## 2. Enregistrement de requêtes

```python
# Ajouter plusieurs requêtes
tracker.add_request(
    request_name="Analyse de sentiment",
    prompt_tokens=25,
    completion_tokens=10,
    model="mistral-tiny",
    temperature=0.7
)

tracker.add_request(
    request_name="Traduction FR->EN",
    prompt_tokens=50,
    completion_tokens=45,
    model="mistral-small",
    metadata={"document": "rapport.pdf", "urgent": True}
)

tracker.add_request(
    request_name="Résumé de document",
    prompt_tokens=200,
    completion_tokens=150,
    model="mistral-medium",
    temperature=0.3
)

print("3 requêtes enregistrées")
```

## 3. Statistiques consolidées

```python
# Statistiques par jour
df_consolidated = tracker.get_consolidated_stats()
df_consolidated
```

```python
# Utilisation totale
total_usage = tracker.get_total_usage()
print(f"Tokens prompt totaux: {total_usage['total_prompt_tokens']}")
print(f"Tokens completion totaux: {total_usage['total_completion_tokens']}")
print(f"Tokens totaux: {total_usage['total_tokens']}")
print(f"Nombre total de requêtes: {total_usage['request_count']}")
```

## 4. Détails par jour

```python
# Détails pour aujourd'hui
from datetime import datetime
today = datetime.now().strftime("%Y-%m-%d")
df_daily = tracker.get_daily_stats(today)
df_daily
```

## 5. Filtrer par modèle

```python
# Requêtes pour mistral-tiny
df_mistral_tiny = tracker.get_requests_by_model("mistral-tiny")
df_mistral_tiny
```

## 6. Visualisation (si matplotlib est installé)

```python
try:
    import matplotlib.pyplot as plt
    
    # Graphique d'utilisation quotidienne
    df = tracker.get_consolidated_stats()
    if not df.empty:
        plt.figure(figsize=(10, 5))
        plt.bar(df['date'], df['total_tokens'], color='skyblue')
        plt.title('Utilisation quotidienne des tokens')
        plt.xlabel('Date')
        plt.ylabel('Tokens totaux')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
except ImportError:
    print("Matplotlib non installé. Installez-le avec: pip install matplotlib")
```

## 7. Export des données

```python
# Exporter vers CSV
tracker.export_to_csv("mon_usage_tokens.csv")
print("Données exportées vers mon_usage_tokens.csv")

# Vérifier que le fichier existe
import os
os.path.exists("mon_usage_tokens.csv")
```

## 8. Calcul de coût (exemple)

```python
# Coût par token (exemple pour Mistral - à adapter selon votre fournisseur)
COST_PER_PROMPT_TOKEN = 0.000002  # $0.000002 par token
COST_PER_COMPLETION_TOKEN = 0.000006  # $0.000006 par token

total = tracker.get_total_usage()
cost = (total['total_prompt_tokens'] * COST_PER_PROMPT_TOKEN + 
        total['total_completion_tokens'] * COST_PER_COMPLETION_TOKEN)

print(f"Coût total estimé : ${cost:.6f}")
```
